### Samples from CSV File  ISAACS - 2D - Gold and Copper

###### Importing Python library

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pygeostat as gs # CCG Geostatistics library
import os
```

### Reading CSV File

```python
path=os.getcwd()
filename=path+"\ISAACS.CSV"
print(filename)
tablaDF = gs.DataFile(flname=filename, dftype='csv', headeronly=False)  #tablaDF=gs.read_csv(filename)
tablaDF.head()
```

```python
gs.location_plot(tablaDF, var='copper')
gs.location_plot(tablaDF, var='gold')
```

```python
gs.scatter_plots(tablaDF, logx=False, logy=False)
```

# Correlation Matrix

```python
data=tablaDF.variables
print("Variables: ",data)
```

```python
data = tablaDF[tablaDF.variables]
data_cor = data.corr()
gs.correlation_matrix_plot(data_cor,  annotation=True) # lower_matrix=True
```

# Additional Resuls from Bivariate Stats
### Coeficiente de Correlacion, Covarianza and Variance Matrix

```python
print("Coeficiente de Correlacion matrix")
data.corr()
```

```python
print("Covariance matrix")
data.cov()
```

```python
print("Variance matrix - Univariate Stats")
print(data.var())
```
